<h3>Available datasets for Sentiment Analysis-related projects</h3>

1. <a href="http://www.cs.jhu.edu/~mdredze/datasets/sentiment/">Multi-Domain Sentiment Dataset</a>

Dataset is composed of product reviews taken from Amazon.com from many product domains like *books*, *dvds*, and *musical instruments*. Reviews contain star ratings ranged from 1 to 5 stars which can be changed to **positive** and **negative** labels respectively. Thus such a dataset is suitable for `binary sentiment classification`. There are also a few options regarding the extent to what the dataset is processed varying from the original data set to thoroughly pre-processed and label-balanced. The data set is relatively small in size. For example, the size of the pre-processed version is less than 33 MB.
2. <a href="http://ai.stanford.edu/~amaas/data/sentiment/">IMDB reviews (Large Movie Review Dataset)</a>

Dataset contains movie reviews for `binary sentiment classification` providing 50,000 reviews overall which are splited into 25,000 samples for training and 25,000 samples for testing. The overall distribution of labels is *balanced* (25k
positive and 25k negative). An additional set of 50,000 unlabeled documents for unsupervised learning is also included. In the
labeled train/test sets, a negative review has a score <= 4 out of 10, and a positive review has a score >= 7 out of 10. Thus reviews with more neutral ratings are excluded. The size of the data set is relatively small (80.2 MB).
3. <a href="https://nlp.stanford.edu/sentiment/code.html">Stanford Sentiment Treebank</a>

The corpus is based on the dataset introduced by Pang and Lee (2005) and consists of 11,855 single sentences extracted from an entertainment review website Rotten Tomatoes dataset. It was parsed with the Stanford parser (Klein and Manning, 2003).
4. <a href="http://help.sentiment140.com/for-students/">Sentiment140</a>

Dataset is a CSV file featuring 160,000 tweets with emoticons removed for `multiclass sentiment classification`. Data file format has 6 fields: the sentiment of the tweet (**positive**, **negative**, and **neutral**),  the tweet ID, the date of the tweet, the query, username, and the text of the tweet. The interesting fact about this data set is that the training data was automatically annotated, in contrast the basic approach is the human annotation. The main idea was based on emoticons, i.e. any tweet with positive emoticons, like :), were positive, and tweets with negative emoticons, like :(, were negative.
5. <a href="https://www.kaggle.com/crowdflower/twitter-airline-sentiment#Tweets.csv">Twitter US Airline Sentiment</a>

Dataset accumulates 14640 tweets on six US airlines such as United, US Airways, American, Southwest, Delta, and Virgin America. The tweets are classified as **positive**, **negative**, and **neutral** followed by categorizing negative reasons (such as "late flight" or "rude service"). The dataset is imbalanced and has missing values. The data is formatted under 15 fields including tweet ID, tweet location, text, airline, its sentiment, username, user timezone, etc. The size of the dataset is quite small (about 1.7 MB) which makes it easy to handle with.
6. <a href="https://archive.ics.uci.edu/ml/datasets/Paper+Reviews">Paper Reviews</a>

Dataset includes paper reviews from an international conference on computing and informatics. Most of the reviews are in Spanish, and some of the reviews are in English. The total number of samples equals 405, and the reviews are assessed based on 5-point scale ('-2': **very negative**, '-1': **negative**, '0': **neutral**, '1': **positive**, '2': **very positive**). The distribution of the original scores is more or less uniform. The dataset is stored in JSON format including 10 attributes about paper ID, review ID, review text, remarks, langunage of review, evaluation score, etc. The dataset has missing values. Suitable for both `regression` and `multiclass sentiment classification` tasks.

As it turned out, the raw text is quite messy for the reviews filled with HTML tags and escape sequences. Therefore before doing sentiment analysis the reviews are needed to be cleaned up.

```python
# import library to handle with regular expressions
import re

# Define regular expressions as a pattern objects
brackets_special_chars_1 = re.compile("[\´\'\"()\[\]\\r\\n\*]")
brackets_special_chars_2 = re.compile("[.;:!?,#$]")
html_dashes = re.compile("(<br\s*/><br\s*/>)|(\-)|(\/)|(\x96)|(\x85)")
white_space = re.compile(r"^\s+", re.MULTILINE)


def cleanup(reviews):
    """
    Search defined patterns within the raw text and substitute them for specified alternatives. 
    """
    reviews = [html_dashes.sub("", line.lower()) for line in reviews]
    reviews = [brackets_special_chars_1.sub("", line) for line in reviews]
    reviews = [brackets_special_chars_2.sub(" ", line) for line in reviews]
    reviews = [white_space.sub(" ", line) for line in reviews]
    reviews = [re.sub(' +', ' ',line) for line in reviews]
    reviews = [review.strip() for review in reviews]
    return reviews

# clean up the reviews by replacing defined regular expressions pattern to specified alternative substrings.
reviews_train_clean = cleanup(reviews_train)
reviews_test_clean = cleanup(reviews_test)

# generate the target vector with respect to the reviews arrangments
targets = [1 if i < 12500 else 0 for i in range(25000)]
```

**Evaluate model's performance and interpret classification results**

```python
def get_performance_metrics(y_true, y_pred):
    """
    Compute classification metrics to evaluate the model's performance in terms of
    accuracy, precision, recall, and F-score
    """
    acc = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f_score = f1_score(y_true, y_pred)
    return acc, precision, recall, f_score


y_pred = def_svm.predict(X_devel1) # make prediction on development(hold out) set
acc, precision, recall, f_score = get_performance_metrics(y_devel1, y_pred)
# Report results in terms of accuracy, precision, recall, and F-score metrics
print('Accuracy : {:.2%}'.format(acc))
print('Precision: {:.2%}'.format(precision))
print('Recall   : {:.2%}'.format(recall))
print('F-score  : {:.2%}'.format(f_score))
```

```python
def print_confusion_matrix(y_true, y_pred, labels):
    """
    Plot a visual representation of a confusion matrix as a heatmap
    """
    cm = confusion_matrix(y_true, y_pred, labels=[1, 0])
    
    cm_df = pd.DataFrame(cm, index=labels, columns=labels)
    confus_matrix = sns.heatmap(cm_df, annot=True, fmt='d', cmap=plt.cm.Blues, cbar=False)
    confus_matrix.set_xlabel('Predicted', fontsize=12, fontweight='bold')
    confus_matrix.set_ylabel('Actual', fontsize=12, fontweight='bold');
```

Make a small qualitative assessment of the mistakes made by the classifier.

```python
from collections import defaultdict


def get_results(dataset, y_true, y_pred):
    """
    Return two dictionaries, first with correctly predicted samples and second with incorrectly predicted samples, given by the
    dataset, ground truth anb predicitons
    """
    if len(y_true) != len(y_pred):
        return -1
    
    correct = defaultdict(list)
    incorrect = defaultdict(list)
    
    for index in range(0, len(y_true)):
        if y_true[index] == y_pred[index]:
            correct[dataset.iloc[index]] = y_pred[index]
        else:
            incorrect[dataset.iloc[index]] = y_pred[index]
    return correct, incorrect


def report_classification_results(prediction_set, top=10):
    """
    Print out examples of either correctly or incorrecly predicted samples depending on a specified set and
    the desired number of examples to be printed
    """
    counter = 0
    for review, prediction in prediction_set.items():
        if counter == top:
            break
        print('{:15s} {}'.format(review, prediction))
        counter += 1

correct, incorrect = get_results(devel_texts1, y_devel1, y_pred)
print('Examples of correctly classified samples:\n')
report_classification_results(correct)
```

**Model improvements**

* <i style="color: blue;">N-gram features</i>

One possible adjustment we can make is to change the way how reviews're vectorized by adding  n-grams (n can be set to 2, 3, or 4) of words features to preserve some of the local ordering information. Using that information we can detect more complex common combinations of words to express negative or positive attitude.

```python
def vectorize_and_train(row_texts, labels, vectorizer, C=0):
    """
    Vectorize the given chunck of text in a way that is corresponds to specified vectorizer, mapped text to
    the learnt vocabulary. Then feed vectorized reviews to the classifer. Allows to specify the regularization term.
    """
    vectorizer.fit(row_texts)
    X_train = vectorizer.transform(row_texts)
    
    classifier = LinearSVC(C=2**C, max_iter=1000, random_state=1234)
    classifier.fit(X_train, labels)
    return vectorizer, classifier

def make_prediction(vectorizer, classifier, test_texts, test_labels):
    """
    Make prediction of the given corpus and labels. Preliminary vectorize this corpus following the passed vectorizer.
    """
    X_test = vectorizer.transform(test_texts)
    pred_classes = classifier.predict(X_test)
    return pred_classes


def predict_evaluate_and_report(vectorizer, classifier, test_texts, test_labels):
    """Report the results of classification of given corpus and its labels, model, and vecotizer
    used to transform the corpus before feeding it to the classifier"""
    X_test = vectorizer.transform(test_texts)
    pred_classes = classifier.predict(X_test)
    acc, precision, recall, f_score = get_performance_metrics(test_labels, pred_classes)
    print('Accuracy: {:.2%}  Precision: {:.2%}  Recall: {:.2%}  F-score: {:.2%}'.format(acc, precision, recall, f_score))
    

bigram_vectorizer =  TfidfVectorizer(tokenizer=space_tokenizer, ngram_range=(1,2))
trigram_vectorizer =  TfidfVectorizer(tokenizer=space_tokenizer, ngram_range=(1,3))
fourgram_vectorizer =  TfidfVectorizer(tokenizer=space_tokenizer, ngram_range=(1,4))

_, bigram_classifier = vectorize_and_train(train_texts1, y_train1, bigram_vectorizer)
_, trigram_classifier = vectorize_and_train(train_texts1, y_train1, trigram_vectorizer)
_, fourgram_classifier = vectorize_and_train(train_texts1, y_train1, fourgram_vectorizer)

print('Bigram', end='  ')
predict_evaluate_and_report(bigram_vectorizer, bigram_classifier, devel_texts1, y_devel1)
print()
print('Trigram', end='  ')
predict_evaluate_and_report(trigram_vectorizer, trigram_classifier, devel_texts1, y_devel1)
print()
print('Fourgram', end='  ')
predict_evaluate_and_report(fourgram_vectorizer, fourgram_classifier, devel_texts1, y_devel1)
```

Additional features represented by 2-grams, 3-grams and 4-grams of tokens have not resulted in any improvements in classifier's performance in terms of accuracy, precision, recall, and F-score. In contrast, such features has worsened the ability of classifer to predict the review sentiment implying that features are overwhelming.

* <i style="color: blue;">Normalization of the TF-IDF features</i>

Another option to change the way how reviews are vectorized is to normalize the TF-IDF features representation. In practice, this procedure tends to increase the classifier's performance. 

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

countvectorizer = CountVectorizer(tokenizer=space_tokenizer)
countvectorizer.fit(train_texts1)

tokens_counts_train = countvectorizer.transform(train_texts1)
tokens_counts_devel = countvectorizer.transform(devel_texts1)

tfidf_transformer = TfidfTransformer()

_, classifier = vectorize_and_train(tokens_counts_train, y_train1, tfidf_transformer)
predict_evaluate_and_report(tfidf_transformer, classifier, tokens_counts_devel, y_devel1)
```

Normalization of the TF-IDF features has not affected drastically the classifer's performance. As it turned out, the reason for no effect is when creating an instance of *TfidfVectorizer* class, one of the parameters is **norm** which stands for the vector length normalization, and its default value is 'l2' that ensures that the sum of squares of vector elements is 1.

* <i style="color: blue;">Tunning of the regularization term</i>

Here we're going to vary regularization parameter `C` for SVM, which controls the cost of errors within the margin area, and measure performance on the develompnet (validation) set.

```python
results = list() # hold the current value of C parameter and the accuracy score corresponding to this value
c_range = np.linspace(-6, 6, 24) # values of the regularization term to test
for C in c_range:
    vect, classifier = vectorize_and_train(train_texts1, y_train1, tfidf_vectorizer, C=C)
    X_test = vect.transform(devel_texts1)
    pred_classes = classifier.predict(X_test)
    acc, precision, recall, f_score = get_performance_metrics(y_devel1, pred_classes)
    results.append({'C': 2**C, 'acc': acc})
    
    print('C: {:6.3f}'.format(2**C), end='  ')
    print('Accuracy: {:.2%}  Precision: {:.2%}  Recall: {:.2%}  F-score: {:.2%}'.format(acc, precision, recall, f_score))
```

Visualize the results of tunning of regularization term to make the results more illustrative in order to more easily identify the optimal value of regularization parameter.

```python
reg_term_tunning = pd.DataFrame(results)
line = reg_term_tunning.plot(x='C', y='acc', logx=True, figsize=(8, 6))
line.set_xscale('log', basex=2);
```

The figure above illustrates the effect of changes in value of regularization term for SVM for classifer's performance on development set in terms of its accuracy score. First, setting a regularization term to small values leads to **underfitting**, then it reaches the *optimal value* for `C` parameter that yields the best performance. Afterwards the accuracy starts dropping as the regularization grows stronger signifying that model's **overfitting**.

```python
best_accuracy = reg_term_tunning['acc'].max()
index = reg_term_tunning['acc'].idxmax()
power_value1 = c_range[index]

print('Accuracy: {:.2%}'.format(best_accuracy))
print('Best C (regularization parameter): 2^{:.2f} ~ {:.2f}'.format(power_value1, 2**power_value1))
```

As a result, we've worked out the optimal value of `C` parameter, $2^{-0.78}$ ($\simeq 0.58$), that maximizes classifier's performance on the development set up to 78.00%

Now it's time to train the classifer on training and development sets merged together and evaluate its performance on test set we hold out in the very beginning when splitting the original dataset into three individual subsets.
